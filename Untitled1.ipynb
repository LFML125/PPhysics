```python
import random
import matplotlib.pyplot as plt

# Dimensiones de la caja
box_width = 100
box_height = 80
radius = 2  # Radio de las partículas
num_particles = 20

# Lista para almacenar las posiciones de las partículas
particles = []

def is_valid_position(x, y, particles, radius):
    """Comprueba si una nueva posición (x, y) está suficientemente lejos de las existentes."""
    for px, py in particles:
        distance = ((x - px)**2 + (y - py)**2)**0.5
        if distance < 2 * radius:  # Mínima distancia entre partículas
            return False
    return True

# Generar partículas aleatorias
while len(particles) < num_particles:
    x = random.uniform(radius, box_width - radius)
    y = random.uniform(radius, box_height - radius)
    if is_valid_position(x, y, particles, radius):
        particles.append((x, y))

# Visualización
fig, ax = plt.subplots()
ax.set_xlim(0, box_width)
ax.set_ylim(0, box_height)
ax.set_aspect('equal')

# Dibujar la caja y las partículas
for x, y in particles:
    circle = plt.Circle((x, y), radius, color='blue', fill=True)
    ax.add_patch(circle)

# Dibujar los bordes de la caja
plt.gca().add_patch(plt.Rectangle((0, 0), box_width, box_height, fill=None, edgecolor='black'))

plt.title("Caja con partículas aleatorias")
plt.show()







import matplotlib.pyplot as plt
import numpy as np

# Dimensiones de la caja
box_width = 100
box_height = 80
num_particles = 20
radius = 2  # Radio de las partículas

# Crear una cuadrícula para distribuir equidistantemente las partículas
rows = int(np.sqrt(num_particles))  # Número de filas de partículas
cols = int(np.ceil(num_particles / rows))  # Número de columnas de partículas

# Espaciado entre partículas
x_spacing = box_width / cols
y_spacing = box_height / rows

# Lista para las posiciones de las partículas
particles = []

# Generar las posiciones
for i in range(rows):
    for j in range(cols):
        if len(particles) < num_particles:
            x = (j + 0.5) * x_spacing  # Centrar en cada celda de la cuadrícula
            y = (i + 0.5) * y_spacing
            particles.append((x, y))

# Visualización
fig, ax = plt.subplots()
ax.set_xlim(0, box_width)
ax.set_ylim(0, box_height)
ax.set_aspect('equal')

# Dibujar la caja y las partículas
for x, y in particles:
    circle = plt.Circle((x, y), radius, color='red', fill=True)
    ax.add_patch(circle)

# Dibujar los bordes de la caja
plt.gca().add_patch(plt.Rectangle((0, 0), box_width, box_height, fill=None, edgecolor='black'))

plt.title("Caja con partículas equidistantes")
plt.show()
```
